###Logistic Regression

    -   Using Linear Regression to represent probabilities may not be the best approach because the output probabilities can go out of range (0,1) - they can be negative or > 1.
    -   Logistic Regression uses a different function to restrict the output to (0,1) - the logistic function.

To fit the model, we use a method called maximum likelihood

The logistic function will always produce an S-shaped curve of this form, and so regardless of the value of X, we will obtain a sensible prediction
